```python
def calculate_age(birth): 
    age = today.year - birth.year - ((today.month, today.day) < (birth.month, birth.day)) 
  
    return age
```

```python
sql_users_query = 'select * from users';

users_df = pd.read_sql_query(sql_users_query, conn)

users_total = len(users_df)
fem_users_total = len(users_df[users_df['gender']=='fem'])
mas_users_total = len(users_df[users_df['gender']=='masc'])

users_df['age'] = users_df['birth'].apply(calculate_age)
users_df['age'].hist(bins=30)
plt.show()
```

```python
print('Total de usu치rios:', users_total)
print('Total de mulheres:', fem_users_total, '->', round((fem_users_total/users_total)*100,2), '%')
print('Total de homens:', mas_users_total, '->', round((mas_users_total/users_total)*100,2), '%')
```

# An치lise dos coment치rios

```python
sql_comments_avg_votes = """
select comments.comment_id,
       comments.content,
       avg(vote)
from comments join votes on comments.comment_id = votes.comment_id
group by (comments.comment_id,comments.content,vote);
"""
comments_content = pd.read_sql_query(sql_comments_avg_votes, conn)
```

```python
def treat_string():
    stop_words = stopwords.words('portuguese')
    lower_chars = lambda comment: comment.lower()
    remove_rt_user = lambda word: re.sub(r'(!* RT @[A-Za-z0-9_]*)|(!*"@[A-Za-z0-9_]*)', '', word)
    remove_user = lambda word: re.sub(r'@[A-Za-z0-9_]*', '', word)
    remove_rt = lambda word: re.sub(r' rt ', '', word)
    remove_special_char = lambda word: re.sub(r'[!.,\-";:~><^%$#@&*\(\)-+=_\'\?]','', word)
    remove_stopwords = lambda comment: ' '.join([word for word in comment.split(' ') if word not in stop_words])

    comments_content['comment'] = comments_content['content'].apply(remove_rt_user)  # remove o inicio '!!!! RT @sjdha'
    comments_content['comment'] = comments_content['content'].apply(lower_chars)   # coloca tudo em minusculo
    comments_content['comment'] = comments_content['content'].apply(remove_user)   # remove os usuarios
    comments_content['comment'] = comments_content['content'].apply(remove_rt)   # remove os rt
    comments_content['comment'] = comments_content['content'].apply(remove_special_char)   # remove os caracteres especiais
    comments_content['comment'] = comments_content['content'].apply(remove_stopwords)      # remove as stopwords
```

```python
treat_string()
sexist_comments = comments_content[comments_content['avg'] >= 0.5]
non_sexist_comments = comments_content[comments_content['avg'] < 0.5]
```

```python
tf_data = CountVectorizer()
tf_sexist = CountVectorizer()
tf_non_sexist = CountVectorizer()

bow_data = tf_data.fit_transform(comments_content['comment'])
bow_sexist = tf_sexist.fit_transform(sexist_comments['comment'])
bow_non_sexist = tf_non_sexist.fit_transform(non_sexist_comments['comment'])
```

```python
voc_data = tf_data.vocabulary_
voc_sexist = tf_sexist.vocabulary_
voc_non_sexist = tf_non_sexist.vocabulary_

voc_data_size = len(voc_data.keys())
voc_sexist_size = len(voc_sexist.keys())
voc_non_sexist = len(voc_non_sexist.keys())
```

```python
matrix_data = bow_data.toarray()
matrix_sexist = bow_sexist.toarray()
matrix_non_sexist = bow_non_sexist.toarray()
```

```python
# pega todas as palavras de cada classe
sexist_words = tf_sexist.get_feature_names()
non_sexist_words = tf_non_sexist.get_feature_names

all_sexist_words = ' '.join([texto for texto in sexist_comments['comment']]).lower()
all_non_sexist_words = ' '.join([texto for texto in non_sexist_comments['comment']]).lower()
```

```python
%matplotlib inline

sexist_cloud = WordCloud(width= 800, height= 500,
                          max_font_size = 110,
                          collocations = False, 
                          max_words=100,
                          background_color="white").generate(all_sexist_words)

non_sexist_cloud = WordCloud(width= 800, height= 500,
                          max_font_size = 110,
                          collocations = False,
                          max_words=100,
                          background_color="white").generate(all_non_sexist_words)
```
