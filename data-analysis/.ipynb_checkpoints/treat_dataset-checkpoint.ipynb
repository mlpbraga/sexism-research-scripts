## Tratamento de tweets

```python
def treat_string():
    stop_words = stopwords.words('english')
    lower_chars = lambda tweet: tweet.lower()
    remove_rt_user = lambda word: re.sub(r'(!* RT @[A-Za-z0-9_]*)|(!*"@[A-Za-z0-9_]*)', '', word)
    remove_user = lambda word: re.sub(r'@[A-Za-z0-9_]*', '', word)
    remove_rt = lambda word: re.sub(r' rt ', '', word)
    remove_special_char = lambda word: re.sub(r'[!.,\-";:~><^%$#@&*\(\)-+=_\'\?]','', word)
    remove_stopwords = lambda tweet: ' '.join([word for word in tweet.split(' ') if word not in stop_words])

    labeled_data['tweet_1'] = labeled_data['tweet'].apply(remove_rt_user)  # remove o inicio '!!!! RT @sjdha'
    labeled_data['tweet_1'] = labeled_data['tweet_1'].apply(lower_chars)   # coloca tudo em minusculo
    labeled_data['tweet_1'] = labeled_data['tweet_1'].apply(remove_user)   # remove os usuarios
    labeled_data['tweet_1'] = labeled_data['tweet_1'].apply(remove_rt)   # remove os rt
    labeled_data['tweet_1'] = labeled_data['tweet_1'].apply(remove_special_char)   # remove os caracteres especiais
    labeled_data['tweet_2'] = labeled_data['tweet_1'].apply(remove_stopwords)      # remove as stopwords
```

## Calculo de Entropias

Entropia de **Shanon**: 

```python
i_shanon_hate = dict({}) # dicionario para guardar entropias com o histograma individual de Ã³dio
doc_hate_id = 0

for line in bow_hate.toarray():
    i_shanon_hate[doc_hate_id] = entropy(line)/log(voc_hate_size)
    doc_hate_id +=1
    
i_shanon_offensive = dict({}) # dicionario para guardar entropias com o histograma individual ofensivo
doc_offensive_id = 0

for line in bow_offensive.toarray():
    i_shanon_offensive[doc_offensive_id] = entropy(line)/log(voc_offensive_size)
    doc_offensive_id +=1

i_shanon_regular = dict({}) # dicionario para guardar entropias com o histograma individual regular
doc_regular_id = 0

for line in bow_regular.toarray():
    i_shanon_regular[doc_regular_id] = entropy(line)/log(voc_regular_size)
    doc_regular_id +=1
```
